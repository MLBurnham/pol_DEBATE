```python
import pandas as pd
import os
```

```python
folders = os.listdir("../data/ezstance/subtaskB/noun_phrase")
```

```python
# Initialize an empty list to store dataframes
dataframes = []

# Loop through each folder
for folder in folders:
    folder = "../data/ezstance/subtaskB/noun_phrase/" + folder
    # List all files in the folder
    files = os.listdir(folder)

    # Loop through files and process train, test, and val files
    for file in files:
        file_path = os.path.join(folder, file)
        if "train" in file.lower() and file.endswith(".csv"):
            train_df = pd.read_csv(file_path)
            train_df['split'] = 'train'
            dataframes.append(train_df)
        elif "test" in file.lower() and file.endswith(".csv"):
            test_df = pd.read_csv(file_path)
            test_df['split'] = 'test'
            dataframes.append(test_df)
        elif "val" in file.lower() and file.endswith(".csv"):
            val_df = pd.read_csv(file_path)
            val_df['split'] = 'val'
            dataframes.append(val_df)

# Combine all the dataframes into a single dataframe
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
df = combined_df.drop_duplicates(subset = ['Text', 'Target 1', 'Ori Text'])
```

```python
df.drop(labels = ['Keywords', 'In Use', 'seen?'], axis = 1, inplace = True)
df.columns = ['text', 'target', 'stance', 'domain', 'ori_text', 'split']
df
```
